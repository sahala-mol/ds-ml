#### Activity 1: Create scatter plot.



```
# 1. Create a scatter plot between the 'SepalLengthCm' & 'SepalWidthCm' columns & Differentiate between the data points of different classes using the 'hue' parameter.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/sahala-mol/ds-ml/main/Iris%20(1).csv')
sns.scatterplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'],hue=df['Species'])
plt.show()
```

```
# 2. Create a scatter plot between the 'PetalLengthCm' & 'PetalWidthCm' columns & Differentiate between the data points of different classes using the 'hue' parameter.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/sahala-mol/ds-ml/main/Iris%20(1).csv')
sns.scatterplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],hue=df['Species'])
plt.show()
```
